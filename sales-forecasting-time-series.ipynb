```python
oil['date']=pd.to_datetime(oil['date'])
```

```python
oil.info()
```

```python
#Finding the mean oil price of each year to take care of the missing values
mean_oil_2013=oil[(oil['date'].dt.year==2013) & (oil['dcoilwtico'].isnull()==False)].mean()
mean_oil_2014=oil[(oil['date'].dt.year==2014) & (oil['dcoilwtico'].isnull()==False)].mean()
mean_oil_2015=oil[(oil['date'].dt.year==2015) & (oil['dcoilwtico'].isnull()==False)].mean()
mean_oil_2016=oil[(oil['date'].dt.year==2016) & (oil['dcoilwtico'].isnull()==False)].mean()
mean_oil_2017=oil[(oil['date'].dt.year==2017) & (oil['dcoilwtico'].isnull()==False)].mean()
```

```python
#populating the mean of the each year for that year's missing values.
import math
i=0
for year,price in zip(oil['date'].dt.year,oil['dcoilwtico']):
    #print(price,year)
    if math.isnan(price)==True and year==2013:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2013)
    elif math.isnan(price)==True and year==2014:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2014)
    elif math.isnan(price)==True and year==2015:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2015)
    elif math.isnan(price)==True and year==2016:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2016)
    elif math.isnan(price)==True and year==2017:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2017)
    i+=1
```

The dataset contains null value as there are certain holidays on which the store was closed and hence there were no transaction details for them. Handling the missing values

```python
trans_fix=X_train[(X_train['transactions'].isnull()==True) & (X_train['sales']>0.0)]
```

```python
trans_fix
```

```python
trans[(trans['date']=='2013-06-19')&(trans['store_nbr']==54)]
```

```python
trans_fix.shape
```

The above line shows that for few non-holidays there are no records in the transaction table, so checking which stores has such missing transaction data on normal days where there are sales data present

```python
agg_trans=trans.groupby(['store_nbr','year',trans['date'].dt.month])['transactions'].mean().reset_index()
agg_trans['transactions']=agg_trans['transactions'].astype(int)
agg_trans
```

```python
agg_trans.info()
```

```python
for year,month,str_nbr,i in zip(trans_fix['date'].dt.year,trans_fix['date'].dt.month,trans_fix['store_nbr'],trans_fix['id']):
    j=0
    #print(i,str_nbr)
    for yr,mon,str1 in zip(agg_trans['year'],agg_trans['date'],agg_trans['store_nbr']):
        if year==yr and month==mon and str_nbr==str1:
            idx=trans_fix.index[trans_fix['id']==i]
            trans_fix.loc[idx,'transactions']=agg_trans.loc[j,'transactions']
        j+=1
```

```python
trans_fix.info()
```

```python
for i,tran in zip(trans_fix['id'],trans_fix['transactions']):
    idx=X_train.index[X_train['id']==i]
    X_train.loc[idx,'transactions']=tran
```

```python
samp=X_train[(X_train['transactions'].isnull()==True) & (X_train['sales']==0.0) & (X_train['store_holiday']==1.0)]
samp
```

```python
def holiday_trans_null(sales,holiday,tran):
    if sales==0.0 and math.isnan(tran)==True: # and holiday==1.0:
        tran=0
    return tran
```

```python
#Setting the transaction count to 0 for all those rows which have 0 sales data and store_holiday indicator as 1
X_train['transactions']=X_train.apply(lambda val: holiday_trans_null(val['sales'],val['store_holiday'],val['transactions']),axis=1)
```
