Some EDA to check the seasonality and product sales over different stores

```python
#Convert trans column date to Datetime object
trans['date']=pd.to_datetime(trans['date'])
trans.info()
```

```python
trans['year']=trans['date'].dt.year
```

```python
import matplotlib.pyplot as plt
#top 10 stores
top10=trans.groupby(['store_nbr','year'],sort=True)['transactions'].sum().reset_index()
top10.sort_values('transactions',ascending=False).head(10)
```

```python
#Checking the transactions of store over years
top10=top10.sort_values('transactions',ascending=False).head(30)
top10.shape
```

```python
#Plotting the yearwise performance of the stores yielding high num of transactions
plt.style.use('seaborn')
top10.pivot(index='year',columns='store_nbr',values='transactions').plot(kind='bar',figsize=(10,6))
plt.xlabel('Year')
plt.ylabel('Transactions')
plt.show()
```

Fewer transactions in 2017, so lets dig deep and see what all stores yielded most count of transactions in 2017

```python
trans_2017=trans[trans['year']==2017]
```

```python
top10_2017=trans_2017.groupby(['store_nbr','year'],sort=True)['transactions'].sum().reset_index()
top10_2017=top10_2017.sort_values('transactions',ascending=False).head(10)
top10_2017
```

```python
#plotting 2017 top 10 rows
top10_2017.pivot(index='year',columns='store_nbr',values='transactions').plot(kind='bar',figsize=(6,4))
plt.show()
```

Before doing further EDA, joining the datasets

```python
oil['date']=pd.to_datetime(oil['date'])
```

```python
oil.info()
```

```python
#Finding the mean oil price of each year to take care of the missing values
mean_oil_2013=oil[(oil['date'].dt.year==2013) & (oil['dcoilwtico'].isnull()==False)].mean()
mean_oil_2014=oil[(oil['date'].dt.year==2014) & (oil['dcoilwtico'].isnull()==False)].mean()
mean_oil_2015=oil[(oil['date'].dt.year==2015) & (oil['dcoilwtico'].isnull()==False)].mean()
mean_oil_2016=oil[(oil['date'].dt.year==2016) & (oil['dcoilwtico'].isnull()==False)].mean()
mean_oil_2017=oil[(oil['date'].dt.year==2017) & (oil['dcoilwtico'].isnull()==False)].mean()
```

```python
#populating the mean of the each year for that year's missing values.
import math
i=0
for year,price in zip(oil['date'].dt.year,oil['dcoilwtico']):
    #print(price,year)
    if math.isnan(price)==True and year==2013:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2013)
    elif math.isnan(price)==True and year==2014:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2014)
    elif math.isnan(price)==True and year==2015:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2015)
    elif math.isnan(price)==True and year==2016:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2016)
    elif math.isnan(price)==True and year==2017:
        oil.loc[i,'dcoilwtico']=float(mean_oil_2017)
    i+=1
```
